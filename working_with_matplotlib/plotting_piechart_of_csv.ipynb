```
import matplotlib.pyplot as plt
import pandas as pd
```

```
df = pd.read_csv('data.csv')

# Plotting Marks by name of student.
plt.pie(df.iloc[0:,3],labels=df.iloc[0:,0])

plt.show()
```

```
# Plotting graph of those who obtained marks more than 50.

x=df[df["marks"]>50]
plt.pie(x.iloc[0:,3], labels=x.iloc[0:,0],autopct="%1.1f%%")

plt.show()
```
