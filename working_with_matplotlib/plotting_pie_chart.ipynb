```
import numpy as np
import matplotlib.pyplot as plt
import time
```

```
x=[4,7,8,9,10]
y=[13,23,5,7,9]
player=['Dhoni','Virat','Harsh']
runs=[100,78,58]
exp=[0.1,0.2,0]   #  To display the data differently from other
```

```
plt.xlabel('x-axis')
plt.ylabel('y-axis')
#plt.bar(x,y,label='pollution')
#plt.bar(player,runs,label='cricketscore')

plt.scatter(x,y)   #  Displays only the points

plt.legend()
plt.grid(color='green')
plt.show()
```

```

plt.pie(runs,labels=player)   #  To display the PIE Chart (data, then label to data)
plt.show()
```

```

plt.pie(runs,labels=player,explode=exp)   #  Explode - display data different from others)
plt.show()
```

```

plt.pie(runs,labels=player,explode=exp,shadow=True)   #  Shadow - DIsplays the shadow to make it 3D
plt.show()
```

```

plt.pie(runs,labels=player,explode=exp,shadow=True,autopct='%1.1f%%')   #  autopct - Displays the percentage of the slice
plt.show()
```
